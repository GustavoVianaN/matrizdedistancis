```python
from spacy.language import Language
import spacy
from bs4 import BeautifulSoup

!wget 'https://en.wikipedia.org/wiki/Natural_language_processing' -O site1.html
!wget 'https://www.datarobot.com/blog/what-is-natural-language-processing-introduction-to-nlp/' -O site2.html
!wget 'https://hbr.org/2022/04/the-power-of-natural-language-processing' -O site3.html
!wget 'https://www.wonderflow.ai/blog/natural-language-processing-examples' -O site4.html
!wget 'https://www.techtarget.com/searchenterpriseai/definition/natural-language-processing-NLP' -O site5.html

def criacaoFuncao(numerosite):
  while numerosite < 5:
    numerosite = numerosite + 1
    text = "".join([a.get_text() for a in BeautifulSoup(open("./site{}.html".format(numerosite), 'r'.format(numerosite)), 'html.parser').find_all('p')])
    nlp = spacy.load("en_core_web_sm")
    nlp.add_pipe("set", before="parser")
    doc = nlp(text)
    array2 = [sent.text for sent in nlp(text).sents]
    array2 = len(array2)
    print("array de sentenças :", [sent.text for sent in doc.sents])
    print(len(text), " palavras - ", array2, " sentenças no array de sentenças \n")

@Language.component("set")
def set(doc):
    for token in doc[:-1]:
        if token.text == "," or token.text == "..." or token.text == "?" or token.text == "!":
            doc[token.i + 1].is_sent_start = True
    return doc
criacaoFuncao(0)


import numpy
import re

def criarSentencas(sentences):
    listadePalavras = []
    for sentence in sentences:
        w = word_extraction(sentence)
        listadePalavras.extend(w)
    l = []
    for i in listadePalavras:
        if i not in l:
            l.append(i)
    l.sort()
    return l

def word_extraction(sentence):
    remover = ['a', "the", "is"]
    listadePalavras = re.sub("[^\w]", " ",  sentence).split()
    limpar = [w.lower() for w in listadePalavras if w not in remover]
    return limpar    
    
resultado = []
allsentences = []
contador = 0 
Primeira = []

while contador < 5:
  contador = contador + 1
  text = "".join([a.get_text() for a in BeautifulSoup(open("./site{}.html".format(contador), 'r'.format(contador)), 'html.parser').find_all('p')])
  nlp = spacy.load("en_core_web_sm")
  nlp.add_pipe("set", before="parser")
  doc = nlp(text)
  Primeira = Primeira + [sent.text for sent in nlp(text).sents]

vocab = criarSentencas(Primeira)

def gerarSentenca(Primeira, vocab, resultado):
  for sentence in Primeira:
      listadePalavras = word_extraction(sentence)
      bag_vector = numpy.zeros(len(vocab))
      for w in listadePalavras:
          for i,word in enumerate(vocab):
              if word == w: 
                  bag_vector[i] += 1
      resultado.append(numpy.array(bag_vector))
  return resultado

nlp=spacy.load("en_core_web_sm")
import pandas as pd
df = pd.DataFrame(gerarSentenca(Primeira, vocab, resultado), columns=vocab)
df

import nltk
import re
nltk.download('punkt')

TGFD = 0
textFinal = " "
while TGFD < 5: 
  TGFD = TGFD + 1
  text = ("".join([a.get_text() for a in BeautifulSoup(open("./site{}.html".format(TGFD), 'r'.format(TGFD)), 'html.parser').find_all('p')])) 
  textFinal = textFinal + text

arquivoData = nltk.sent_tokenize(textFinal)
for i in range (len(arquivoData)):
    arquivoData[i] = arquivoData[i].lower()
    arquivoData[i] = re.sub(r'W', ' ', arquivoData[i])
    arquivoData[i] = re.sub(r's+', ' ', arquivoData[i]) 
for i in range (len(arquivoData)):
    arquivoData[i] = arquivoData[i].lower()
    arquivoData[i] = re.sub(r'W', ' ', arquivoData[i]) 
    arquivoData[i] = re.sub(r's+', ' ', arquivoData[i])
ccontador = {}
for data in arquivoData:
    words = nltk.word_tokenize(data)
    for word in words:
        if word not in ccontador.keys():
            ccontador[word] = 1
        else:
            ccontador[word] += 1
import heapq
frequenciPalavras = heapq.nlargest(50,ccontador, key=ccontador.get)
import numpy as np 
palavrasTTTT = {}
for word in frequenciPalavras:
    doc_count = 0 
    for data in arquivoData:
        if word in nltk.word_tokenize(data):
            doc_count += 1
    palavrasTTTT[word] = np.log((len(arquivoData)/doc_count)+1)
matrizTdfd = {}
for word in frequenciPalavras:
    doc_tf = []
    for data in arquivoData:
        frequency = 0
        for w in nltk.word_tokenize(data):
            if w == word:
                frequency += 1
        palavrasTF = frequency/len(nltk.word_tokenize(data))
        doc_tf.append(palavrasTF)
    matrizTdfd[word] = doc_tf
printTfIDF = []
for word in matrizTdfd.keys():
    tfidf = []
    for value in matrizTdfd[word]:
        score = value * palavrasTTTT[word]
        tfidf.append(score)
    printTfIDF.append(tfidf)
print("")
data_df = pd.DataFrame(printTfIDF)
print("The DataFrame is:")
print(data_df)
```
